```
# Canada X and Y values from the csv file
Canada_x_Demographic = df_GDPperCapita.loc[df_GDPperCapita['Country Name'] == 'Canada', '1999':'2020'].values.tolist()[0]
Canada_y_Population = df_populationGrowth.loc[df_populationGrowth['Country Name'] == 'Canada', '1999':'2020'].values.tolist()[0]

# China X and Y values from the csv file
China_x_Demographic = df_GDPperCapita.loc[df_GDPperCapita['Country Name'] == 'China', '1999':'2020'].values.tolist()[0]
China_y_Population = df_populationGrowth.loc[df_populationGrowth['Country Name'] == 'China', '1999':'2020'].values.tolist()[0]

# US X and Y values from the csv file
US_x_Demographic = df_GDPperCapita.loc[df_GDPperCapita['Country Name'] == 'United States', '1999':'2020'].values.tolist()[0]
US_y_Population = df_populationGrowth.loc[df_populationGrowth['Country Name'] == 'United States', '1999':'2020'].values.tolist()[0]

# France X and Y values from the csv file
France_x_Demographic = df_GDPperCapita.loc[df_GDPperCapita['Country Name'] == 'France', '1999':'2020'].values.tolist()[0]
France_y_Population = df_populationGrowth.loc[df_populationGrowth['Country Name'] == 'France', '1999':'2020'].values.tolist()[0]
```

```
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Define the degrees of the polynomial regression for each country
degree_france = 2
degree_us = 1
degree_china = 3
degree_canada = 3

# Fit polynomial regression for each country
poly_france = PolynomialFeatures(degree=degree_france, include_bias=False)
poly_us = PolynomialFeatures(degree=degree_us, include_bias=False)
poly_china = PolynomialFeatures(degree=degree_china, include_bias=False)
poly_canada = PolynomialFeatures(degree=degree_canada, include_bias=False)

# France
x = np.array(list(range(1999, 2021))).reshape(-1, 1)
y_france = np.array(France_y_Population).reshape(-1, 1)
x_future = np.array(list(range(1999, 2034))).reshape(-1, 1)
poly_features_france = poly_france.fit_transform(x)
poly_future_france = poly_france.transform(x_future)
model_france = LinearRegression()
model_france.fit(poly_features_france, y_france)
y_predicted_france = model_france.predict(poly_features_france)
y_predicted_future_france = model_france.predict(poly_future_france)

# Repeat the above steps for other countries (United States, China, Canada)

# United States
y_us = np.array(US_y_Population).reshape(-1, 1)
poly_features_us = poly_us.fit_transform(x)
poly_future_us = poly_us.transform(x_future)
model_us = LinearRegression()
model_us.fit(poly_features_us, y_us)
y_predicted_us = model_us.predict(poly_features_us)
y_predicted_future_us = model_us.predict(poly_future_us)

# China
y_china = np.array(China_y_Population).reshape(-1, 1)
poly_features_china = poly_china.fit_transform(x)
poly_future_china = poly_china.transform(x_future)
model_china = LinearRegression()
model_china.fit(poly_features_china, y_china)
y_predicted_china = model_china.predict(poly_features_china)
y_predicted_future_china = model_china.predict(poly_future_china)

# Canada
y_canada = np.array(Canada_y_Population).reshape(-1, 1)
poly_features_canada = poly_canada.fit_transform(x)
poly_future_canada = poly_canada.transform(x_future)
model_canada = LinearRegression()
model_canada.fit(poly_features_canada, y_canada)
y_predicted_canada = model_canada.predict(poly_features_canada)
y_predicted_future_canada = model_canada.predict(poly_future_canada)

# Plot the regression lines and predicted values for each country
plt.figure(figsize=(10, 6))

# France
sns.regplot(x=x.flatten(), y=y_france.flatten(), order=degree_france, label='France Regression')
sns.regplot(x=x_future.flatten(), y=y_predicted_future_france.flatten(), ci=95, scatter=False, order=degree_france, label='France Predicted', color='blue')

# United States
sns.regplot(x=x.flatten(), y=y_us.flatten(), order=degree_us, label='United States Regression')
sns.regplot(x=x_future.flatten(), y=y_predicted_future_us.flatten(), ci=95, scatter=False, order=degree_us, label='United States Predicted', color='orange')

# China
sns.regplot(x=x.flatten(), y=y_china.flatten(), order=degree_china, label='China Regression')
sns.regplot(x=x_future.flatten(), y=y_predicted_future_china.flatten(), ci=95, scatter=False, order=degree_china, label='China Predicted', color='green')

# Canada
sns.regplot(x=x.flatten(), y=y_canada.flatten(), order=degree_canada, label='Canada Regression')
sns.regplot(x=x_future.flatten(), y=y_predicted_future_canada.flatten(), ci=95, scatter=False, order=degree_canada, label='Canada Predicted', color='red')

# Decorate the plot
plt.xlabel('Year')
plt.ylabel('Population Growth (%)')
plt.title('Population Growth over Time')
plt.legend(loc='lower left')

# Display the graph
plt.show()
```
