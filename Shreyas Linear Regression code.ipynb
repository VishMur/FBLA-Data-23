```
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression

data = pd.read_csv('Total Population.csv')

y_data = data.loc[data['Country Name'] == 'Nigeria', '1960':'2020'].values.tolist()[0]
x_data = list(range(1960, 2021))

# Fit a linear regression model to the data
model = LinearRegression().fit(pd.DataFrame(x_data), y_data)

# Make a prediction for the year 2030
x_predict = [[2030]]
y_predict = model.predict(x_predict)

# Plot the original data and the linear regression line
sns.regplot(x=x_data, y=y_data, color='purple')
plt.plot(x_predict, y_predict, 'ro')

plt.show()
```
