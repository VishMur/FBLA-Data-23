```
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
#Pandas reads the Total Population dataset
data = pd.read_csv('Total Population.csv')

#This takes out a specific countries population data from 1960-2020: Y-value is Population, X-value is Time
y_data = data.loc[data['Country Name'] == 'India', '1960':'2020'].values.tolist()[0]
x_data = np.array(list(range(1960, 2021)))

#Creates the features of the polynomial graph
features = PolynomialFeatures(degree=2)
x_poly = features.fit_transform(x_data.reshape(-1, 1))

#This creates the line best fit to the data
model = LinearRegression().fit(x_poly, y_data)

#The next code is the prediction

#The X-value changes to 2030 at its highest X-value
x_new = np.array([2022, 2023, 2024, 2025, 2026, 2027, 2028, 2029, 2030]).reshape(-1, 1)

#The X-value uses the features created before to predict the future new X-value
x_new_poly = features.transform(x_new)

#We then use that to predict the new Y-value
y_new = model.predict(x_new_poly)


#The next code is to create the graph

plt.scatter(x_data, y_data)

# plot the polynomial regression line
plt.plot(x_new, y_new, color='red')
#plt.plot(x_data, y_data, color='blue')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Polynomial Regression for India Population')

plt.show()
```
